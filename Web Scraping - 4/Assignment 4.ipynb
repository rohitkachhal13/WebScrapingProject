### Q.1 Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos

I need to find following details:Rank,Name,Artist,Upload date,Views

```python
# Connecting with web driver
driver = webdriver.Chrome(r'E:/DATA TRAINED/Sofrware/chromedriver_win32/chromedriver.exe')
```

```python
# Open url by driver
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
# Maximize the window
driver.maximize_window()
```

```python
# Creating empty list
rank = []
name = []
artist = []
views = []
upload_date = []

# Scraping the rank
search_rank = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
for i in search_rank:
    rank.append(i.text)

# Scraping the name
search_name = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
for j in search_name:
    name.append(j.text)

# Scraping the artist
search_artist = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[3]')
for k in search_artist:
    artist.append(k.text)
    
# Scraping the views
search_view = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[4]')
for l in search_view:
    views.append(l.text)

# Scraping the upload date
search_upload_date = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[5]')
for m in search_upload_date:
    upload_date.append(m.text)
```

```python
# Printing the length of each list
print(len(rank),len(name),len(artist),len(views),len(upload_date))
```

```python
# Creadting empty data frame
df_wiki = pd.DataFrame({})

# Adding column & their respective data into data frame
df_wiki['Rank'] = rank
df_wiki['Name Of Video'] = name
df_wiki['Artist'] = artist
df_wiki['Views'] = views
df_wiki['Upload Data'] = upload_date
```

```python
# Displaying the result
df_wiki
```

```python
# Fine tunning the data frame
df_wiki['Rank'] = df_wiki['Rank'].str.partition('.')[0]
df_wiki['Name Of Video'] = df_wiki['Name Of Video'].str.partition('[')[0]
df_wiki
```

```python
# Closing the current driver
driver.close()
```

```python
# Clicking on 'Treanding'
click_trending = driver.find_element_by_xpath('//ul[@class="d-lg-flex list-style-none"]/li[4]/details/div/ul/li[5]')
click_trending.click()
```

```python
# Getting the url of all repositories into empty list
url_repositories = []
for i in driver.find_elements_by_xpath('//h1[@class="h3 lh-condensed"]/a'):
    url_repositories.append(i.get_attribute('href'))
url_repositories    
```

```python
# Creating empty list 
title = []
description = []
count = []
language = []
```

```python
# Scraping the title
for i in driver.find_elements_by_xpath('//h1[@class="h3 lh-condensed"]'):
    title.append(i.text)
```

```python
for i in url_repositories:
    driver.get(i)
    time.sleep(3)
    
    # Scraping description
    try :
        description_re = driver.find_element_by_xpath('//p[@class="f4 my-3"]')
        description.append(description_re.text)
    except :
        description.append('-')
        
    # Scraping count
    try :
        count_re = driver.find_element_by_xpath('//*[contains(text(),"Contributors")]')
        count.append(count_re.text)
    except :
        count.append('-')
        
    # Scraping language
    lang = []
    try :
        language_re = driver.find_element_by_xpath('//*[contains(text(),"Languages")]/following::div[1]/following::ul[1]')
        lang.append(language_re.text.replace('\n',''))
        language.append(lang)
    except :
        language.append('-')
```

```python
# Creating empty list
dataset_name = []
dataset_type = []
task = []
attribute_type = []
no_of_instance = []
no_of_attribute = []
uci_year = []
```

```python
# Scraping data set name
for i in driver.find_elements_by_xpath('//table[@cellpadding="3"]/tbody/tr/td[2]/table[2]/tbody/tr[1]/following::tr/td/table/tbody/tr/td[2]/p/b/a'):
    dataset_name.append(i.text)
    
# Scraping dataset type
for j in driver.find_elements_by_xpath('//table[@cellpadding="3"]/tbody/tr/td[2]/table[2]/tbody/tr[1]/following::tr/td/table/tbody/tr/td[2]/p/b/a/following::td[1]/p'):
    dataset_type.append(j.text)
    
# Scraping task    
for k in driver.find_elements_by_xpath('//table[@cellpadding="3"]/tbody/tr/td[2]/table[2]/tbody/tr[1]/following::tr/td/table/tbody/tr/td[2]/p/b/a/following::td[2]/p'):
    task.append(k.text)

# Scraping attribute type
for l in driver.find_elements_by_xpath('//table[@cellpadding="3"]/tbody/tr/td[2]/table[2]/tbody/tr[1]/following::tr/td/table/tbody/tr/td[2]/p/b/a/following::td[3]/p'):
    attribute_type.append(l.text)
    
# Scraping no of instance
for m in driver.find_elements_by_xpath('//table[@cellpadding="3"]/tbody/tr/td[2]/table[2]/tbody/tr[1]/following::tr/td/table/tbody/tr/td[2]/p/b/a/following::td[4]/p'):
    no_of_instance.append(m.text)
    
# Scraping no of attribute
for n in driver.find_elements_by_xpath('//table[@cellpadding="3"]/tbody/tr/td[2]/table[2]/tbody/tr[1]/following::tr/td/table/tbody/tr/td[2]/p/b/a/following::td[5]/p'):
    no_of_attribute.append(n.text)
    
# Scraping year
for o in driver.find_elements_by_xpath('//table[@cellpadding="3"]/tbody/tr/td[2]/table[2]/tbody/tr[1]/following::tr/td/table/tbody/tr/td[2]/p/b/a/following::td[6]/p'):
    uci_year.append(o.text)   
```

```python
# Printing length of each list
print(len(dataset_name),len(dataset_type),len(task),len(attribute_type),len(no_of_instance),len(no_of_attribute),len(uci_year))
```

```python
# Creadting empty data frame
df_uci = pd.DataFrame({})

# Adding column & their respective data into data frame
df_uci['Dataset Name'] = dataset_name
df_uci['Dataset Type'] = dataset_type
df_uci['Task'] = task
df_uci['Attribute Type'] = attribute_type
df_uci['No Of Instance'] = no_of_instance
df_uci['No Of Attribute'] = no_of_attribute
df_uci['Year'] = uci_year
```
