### Q.2 Now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
# Creating empty list for url
url_amazon = []

# Searching & store the product url
for i in driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']/a"):
    url_amazon.append(i.get_attribute('href'))

# Goto the second page    
next_button_2 = driver.find_element_by_xpath("//div[@class='s-pagination-strip']/a[3]|//li[@class='a-last']|//span[@class='s-pagination-strip']/a[3]")
next_button_2.click()
time.sleep(3)

# Searching & store the product url
for i in driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']/a"):
    url_amazon.append(i.get_attribute('href')) 

# Goto the third page    
next_button_3 = driver.find_element_by_xpath("//div[@class='s-pagination-strip']/a[4]|//li[@class='a-last']|//span[@class='s-pagination-strip']/a[4]")
next_button_3.click()
time.sleep(3)

# Searching & store the product url
for i in driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']/a"):
    url_amazon.append(i.get_attribute('href')) 
```

```python
# printing the length of url
print(len(url_amazon))
```

```python
# printing the url list
url_amazon
```

```python
# Creating empty list
brand_name = []
name_of_product = []
price = []
return_exchange = []
expected_delivery = []
availability = []
product_url = []
```

```python
for i in url_amazon:
    driver.get(i)
    time.sleep(2)
    
    # Fetch the brand name
    try :
        brands=driver.find_element_by_xpath("//div[@id='productOverview_feature_div']/div/table/tbody/tr/td/span[contains(text(),'Brand')]//following::td[1]/span[1]")#scraping brands name by class name='_2WkVRV'
        brand_name.append(brands.text)#appending the text in Brand list
    except :
        brand_name.append('-')
    
    # Fetch the product name
    try :
        product_name=driver.find_element_by_xpath("//h1[@class='a-size-large a-spacing-none']")
        name_of_product.append(product_name.text)
    except :
        name_of_product.append('-')
    
    # Fetch the price
    try :
        prices=driver.find_element_by_xpath("//td[@class='a-span12']")# scraping the price from the xpath
        price.append(prices.text)
    except :
        price.append('-')
    
    # Fetech the return/exchange
    try :
        re_ex=driver.find_element_by_xpath("//div[@id='RETURNS_POLICY']")
        return_exchange.append(re_ex.text)
    except :
        return_exchange.append('-')
        
    # Fetch the expected delivery
    try :
        delivery=driver.find_element_by_xpath("//div[@id='ddmDeliveryMessage']")
        expected_delivery.append(delivery.text)
    except :
        expected_delivery.append('-')
        
    # Fetch the availability
    try :
        stock=driver.find_element_by_xpath("//div[@id='availability']")
        availability.append(stock.text)
    except :    
        availability.append('-')
    
    # Fetch product url
    try :
        purl=driver.find_element_by_xpath("//link[@rel='canonical']")
        product_url.append(purl.get_attribute('href'))
    except :
        product_url.append('-')
```

```python
# printing the length of each list
print(len(brand_name),len(name_of_product),len(price),len(return_exchange),len(expected_delivery),len(availability),len(product_url))
```

```python
# creating the data frame
df = pd.DataFrame({})

# adding the column & their respective value
df['Brand Name'] = brand_name
df['Name Of Product'] = name_of_product
df['Price'] = price
df['Return/Exchange'] = return_exchange
df['Expected Delivery'] = expected_delivery
df['Availability'] = availability
df['Product Url'] = product_url
```

```python
# displaying the data frame
df
```

```python
# Saving the result into CSV
df.to_csv('amazon_search_result.csv')
```

```python
# Closing the driver
driver.close()
```

```python
# Scroll the comments till last
default_h = 0
while True:
        height = driver.execute_script("""
                function getActualHeight() {
                    return Math.max(
                        Math.max(document.body.scrollHeight, document.documentElement.scrollHeight),
                        Math.max(document.body.offsetHeight, document.documentElement.offsetHeight),
                        Math.max(document.body.clientHeight, document.documentElement.clientHeight)
                    );
                }
                return getActualHeight();
            """)
        driver.execute_script(f"window.scrollTo({default_h},{default_h + 200})")
        # fix the time sleep value according to your network connection
        time.sleep(1)
        default_h +=200  
        if default_h >= height:
            break
```
